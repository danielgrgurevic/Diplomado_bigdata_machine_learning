Acción: Comparar.

**Visualización elegida**: Gráfico de Líneas Múltiples.

**Justificación de la visualización**: El gráfico de líneas múltiples permite comparar la evolución temporal del número de infectados en diferentes países.
```python
# V03: Visualización que permita ver el total de infectados en el tiempo por país (fecha).

v03 = pd.read_csv("/content/drive/MyDrive/Data science/Diplomado/Visualización en la era del big data/M3/V03.csv")
v03.head()
```

```python
# Completar con visualización


# Transformar los datos para que estén en formato largo
v03_long = pd.melt(v03, id_vars=['Pais'], var_name='Año', value_name='Contagiados')

# Visualizar las primeras filas del DataFrame transformado
v03_long.head()

# Crear el gráfico de líneas múltiples
plt.figure(figsize=(12, 8))
for country in v03_long['Pais'].unique():
    country_data = v03_long[v03_long['Pais'] == country]
    plt.plot(country_data['Año'], country_data['Contagiados'], marker='o', linestyle='-', label=country)

plt.xlabel('Año')
plt.ylabel('Número de Contagiados')
plt.title('Evolución del Total de Infectados en el Tiempo por País')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# V05: Dos visualizaciones (uno por hemisferio) que permita visualizar la cantidad de fallecimientos por mes y año.
#      De modo que permita determinar si hubo algunos meses en particular donde aumentó la cantidad de fallecidos o
#      siempre se mantuvo de forma homogénea

v05n = pd.read_csv("/content/drive/MyDrive/Data science/Diplomado/Visualización en la era del big data/M3/V05-hem-norte.csv")
v05n.head()
```

```python
v05s = pd.read_csv("/content/drive/MyDrive/Data science/Diplomado/Visualización en la era del big data/M3/V05-hem-sur.csv")
v05s.head()
```

```python
# Completar con visualización

# Transformar los datos para que estén en formato largo
v05n['Fecha'] = v05n['Año'].astype(str) + '-' + v05n['Mes'].str.split('-').str[0]
v05s['Fecha'] = v05s['Año'].astype(str) + '-' + v05s['Mes'].str.split('-').str[0]

# Convertir la columna Fecha a datetime
v05n['Fecha'] = pd.to_datetime(v05n['Fecha'], format='%Y-%m')
v05s['Fecha'] = pd.to_datetime(v05s['Fecha'], format='%Y-%m')

# Ordenar los datos por la columna Fecha
v05n = v05n.sort_values(by='Fecha')
v05s = v05s.sort_values(by='Fecha')

# Visualización para el hemisferio norte
plt.figure(figsize=(12, 8))
plt.plot(v05n['Fecha'], v05n['Fallecidos'], marker='o', linestyle='-', label='Hemisferio Norte', color='b')
plt.xlabel('Fecha')
plt.ylabel('Número de Fallecidos')
plt.title('Evolución de Fallecimientos en el Hemisferio Norte')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Completar con visualización

# Visualización para el hemisferio sur
plt.figure(figsize=(12, 8))
plt.plot(v05s['Fecha'], v05s['Fallecidos'], marker='o', linestyle='-', label='Hemisferio Sur', color='r')
plt.xlabel('Fecha')
plt.ylabel('Número de Fallecidos')
plt.title('Evolución de Fallecimientos en el Hemisferio Sur')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Gráfico de Dispersión (Scatter Plot) con Colores Segmentados por Síntomas

**Justificación**: El gráfico de dispersión permite explorar la correlación entre pares de atributos como edad y peso, altura y edad, y altura y peso.
```python
# V07: Una visualización que permita apreciar distribución y potencialmente correlación entre
#      pares de atributos de dataset: edad, peso y altura; con categorización por tipos de
#      síntomas presentados.

v07 = pd.read_csv("/content/drive/MyDrive/Data science/Diplomado/Visualización en la era del big data/M3/V07.csv")
v07.head()
```

```python
# Completar con visualización.


# Visualización 1: Edad vs Peso por Tipo de Síntomas
plt.figure(figsize=(12, 6))
sns.scatterplot(data=v07, x='Edad', y='Peso', hue='Sintomas', palette='viridis', alpha=0.7)
plt.title('Edad vs Peso por Tipo de Síntomas')
plt.xlabel('Edad')
plt.ylabel('Peso')
plt.legend(title='Síntomas')
plt.show()

# Visualización 2: Altura vs Edad por Tipo de Síntomas
plt.figure(figsize=(12, 6))
sns.scatterplot(data=v07, x='Altura', y='Edad', hue='Sintomas', palette='viridis', alpha=0.7)
plt.title('Altura vs Edad por Tipo de Síntomas')
plt.xlabel('Altura')
plt.ylabel('Edad')
plt.legend(title='Síntomas')
plt.show()

# Visualización 3: Altura vs Peso por Tipo de Síntomas
plt.figure(figsize=(12, 6))
sns.scatterplot(data=v07, x='Altura', y='Peso', hue='Sintomas', palette='viridis', alpha=0.7)
plt.title('Altura vs Peso por Tipo de Síntomas')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.legend(title='Síntomas')
plt.show()
```
